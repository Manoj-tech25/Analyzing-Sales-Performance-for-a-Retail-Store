##Data cleaning

```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv("OnlineRetail.csv", encoding="ISO-8859-1")
```

```python
df
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
df
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
df
```

```python
# Convert data types
df.loc[:, 'InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.loc[:, 'TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
print(df.info())
```

```python
df
```

Exploratory Data Analysis (EDA)

```python
# Regional analysis
top_Country = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
```

```python
print("top_Country:\n", top_Country)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
```

```python
if df['InvoiceDate'].isnull().any():
    print("Some dates could not be converted. Check for invalid values.")
    df = df.dropna(subset=['InvoiceDate']) 
```

```python
# Monthly sales trend
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_sales = df.groupby('Month')['TotalPrice'].sum()
```

```python
print("Monthly Sales:\n", monthly_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line chart - Monthly sales trend -- Simple
monthly_sales.plot(kind='bar', color='green', title='Monthly Sales Trend')
plt.show()
```

```python
# Bar chart - Regional sales -- simple
top_Country.plot(kind='line', marker='o', color='orange', title='Sales by Region')
plt.show()
```

```python
# Plotting Top 10 Countries by Sales
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=top_Country.values, y=top_Country.index, hue=top_Country.index,  palette="viridis",dodge=False, legend=False)
plt.title("Top 10 Countries by Total Sales", fontsize=16)
plt.xlabel("Total Sales (in currency)", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.show()
```

```python
# Plotting Monthly Sales Trends
plt.figure(figsize=(14, 6))
monthly_sales.plot(kind='line', marker='o', color='dodgerblue', linewidth=2)
plt.title("Monthly Sales Trends", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales (in currency)", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# EDA: Customer Purchasing Patterns

# Total spending and frequency per customer
customer_data = df.groupby('CustomerID').agg(
    TotalSpent=('TotalPrice', 'sum'),
    PurchaseFrequency=('InvoiceNo', 'nunique')
).sort_values(by='TotalSpent', ascending=False)

# Top 10 High-Value Customers by Total Spending
top_customers = customer_data.head(10)

# Plotting High-Value Customers
plt.figure(figsize=(12, 4))
sns.barplot(x=top_customers.TotalSpent, y=top_customers.index,  hue=top_Country.index, palette="coolwarm",dodge=False, legend=False)
plt.title("Top 10 High-Value Customers by Total Spending", fontsize=16)
plt.xlabel("Total Spending (in currency)", fontsize=12)
plt.ylabel("Customer ID", fontsize=12)
plt.show()
```

```python
# EDA: Product Popularity

# Top 10 Most Purchased Products by Quantity
product_popularity = df.groupby('Description').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalRevenue=('TotalPrice', 'sum')
).sort_values(by='TotalQuantity', ascending=False).head(10)

# Plotting Most Purchased Products
plt.figure(figsize=(12, 4))
sns.barplot(x=product_popularity.TotalQuantity, y=product_popularity.index, hue=top_Country.index, palette="crest",dodge=False, legend=False)
plt.title("Top 10 Most Purchased Products by Quantity", fontsize=16)
plt.xlabel("Total Quantity Purchased", fontsize=12)
plt.ylabel("Product Description", fontsize=12)
plt.show()
```

```python
# Top 10 Revenue-Generating Products
top_revenue_products = product_popularity.sort_values(by='TotalRevenue', ascending=False).head(10)

# Plotting Top Revenue-Generating Products
plt.figure(figsize=(12, 4))
sns.barplot(x=top_revenue_products.TotalRevenue, y=top_revenue_products.index,hue=top_Country.index, palette="mako",dodge=False, legend=False)
plt.title("Top 10 Revenue-Generating Products", fontsize=16)
plt.xlabel("Total Revenue (in currency)", fontsize=12)
plt.ylabel("Product Description", fontsize=12)
plt.show()
```
